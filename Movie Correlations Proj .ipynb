```python
# importing the packages to be used 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# reading the data /Users/davidyang/Downloads
df = pd.read_csv('/Users/davidyang/Downloads/movies.csv')
```

```python
#viewing dataset

df
```

```python
#checking dataset for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col]. isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#check data types for our columns

df.dtypes
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.drop_duplicates()
```

```python
#Budget vs gross scatter plot

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget Versus Gross Earnings')
plt.xlabel('Gross')
plt.ylabel('Budget')
plt.show
```

```python
# Plot Budget vs gross on scatter

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "green"}, line_kws={"color":"blue"})
```

```python
sns.regplot(x='score', y='gross', data=df)
```

```python
df.corr(numeric_only=True)
```

```python
correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Columns')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
           
plt.show
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized['name'] = df_numerized['name'].cat.codes
df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Columns')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
           
plt.show
```

```python
df_numerized.corr()
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
